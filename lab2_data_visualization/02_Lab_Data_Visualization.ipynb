## Task 1.3 Filtering and Sorting Data:

Filter the dataset to show all games played by the team with id CLE. Then, sort the games by the date.

# 2. Data Visualization 

## Task 1.2 Line Plot:

Create a line plot showing the Elo rating progression (elo_i column) of the LAL team from 1990 to 1998.


```python
lal_games = df[(df['team_id'] == 'LAL') & (df['year_id'] >= 1990) & (df['year_id'] <= 1998)]
plt.plot(lal_games['year_id'], lal_games['elo_i'])

plt.title('LAL Elo Rating (1990-1998)')
plt.xlabel('Date')
plt.ylabel('Elo Rating')
plt.show()
```

## Task 2.2 Bar Plot:

Create a bar chart comparing the total number of games played by the top 5 teams with the most wins.


```python
df_wins = df[df['game_result'] == 'W'].groupby('fran_id').size().reset_index(name='wins')
df_total_games = df.groupby('fran_id').size().reset_index(name='total_games')
df_combined = pd.merge(df_wins, df_total_games, on='fran_id')

top_teams = df_combined.sort_values(by='total_games', ascending=False)
top_5_teams = top_teams[:5]

# write a code to visuzlize data
plt.bar(top_5_teams['fran_id'], top_5_teams['total_games'])
plt.title('Top 5 Teams by Number of Games Played')
plt.xlabel('Teams')
plt.ylabel('Number of Games')
plt.show()
```

## Task 2.3 Scatter Plot: Relationship between Pre-game and Post-game Elo Ratings
Task: Create a scatter plot to visualize the relationship between a team's pre-game (elo_i, opp_elo_i) and post-game (elo_n, opp_elo_n) Elo ratings. 


```python
fig = plt.figure(figsize=(10, 5))
fig.suptitle("Pre-game vs Post-game Elo Ratings")

plt.subplot(1, 2, 1)
plt.scatter(df['elo_i'], df['elo_n'], label='Team elo', s=5, color='blue')
plt.legend()

plt.xlabel('Elo Pre-game')
plt.ylabel('Elo Post-game')

plt.subplot(1, 2, 2)
plt.scatter(df['opp_elo_i'], df['opp_elo_n'], label='Opposite team elo', s=5, color='red')
plt.legend()

plt.xlabel('Elo Pre-game')
plt.ylabel('Elo Post-game')

plt.tight_layout()
plt.show()
```

#  Task 2.4 Pie Chart: Distribution of Game Outcomes by Winning Team
Task: Create a pie chart that visualizes the distribution of wins among the top 5 teams in the dataset.
Analysis: The pie chart will show the proportion of wins among the top 5 teams. 

```python
df_wins = df[df['game_result'] == 'W'].groupby('fran_id').size().reset_index(name='wins')
df_wins = df_wins.sort_values(by='wins', ascending=False)
top_5_winning_teams = df_wins[:5]

plt.pie(top_5_winning_teams['wins'], labels=top_5_winning_teams['fran_id'])
plt.title('Top 5 Teams by Win Distribution')
plt.show()
```

#  Task 2.5 Histogram: Distribution of Game Durations
Task: Create a histogram showing the distribution of game durations (in minutes).


```python
df_pts = df.groupby('team_id')['pts'].sum().reset_index()
plt.hist(df_pts['pts'])
plt.title('Distribution of PTS')
plt.xlabel('PTS')
plt.ylabel('Number of teams')
plt.show()
```

#  Task 2.6 Stacked Bar Plot: Wins by Teams in Different Eras (Decades)
Task: Create a stacked bar plot showing the number of wins by teams, grouped by different decades (1970s, 1980s, 1990s, 2000s, 2010s, etc.).


```python
df['decade'] = (df['year_id'] // 10 * 10).astype(str)+'s'

df_wins = df[df['game_result'] == 'W'].groupby('fran_id').size().reset_index(name='wins')
df_wins = df_wins.sort_values(by='wins', ascending=False)

top_teams = df_wins[:12]['fran_id'].values

filtered_df = df[df['fran_id'].isin(top_teams)]

wins_per_decade = filtered_df.groupby(['decade', 'fran_id']).size().unstack()

# Stacked bar plot
# write a code to visualize dataframe
wins_per_decade.plot(kind='bar', stacked=True, figsize=(10, 7))

plt.title('Wins by Team Across Different Decades')
plt.xlabel('Decade')
plt.ylabel('Number of Wins')
plt.tight_layout()

plt.show()
```

```python
#  Which team won more matches against each other, Lakers or Celtics?

L_vs_C = df[((df['fran_id'] == 'Lakers') & (df['opp_fran'] == 'Celtics')) | ((df['fran_id'] == 'Celtics') & (df['opp_fran'] == 'Lakers'))]

wins = L_vs_C[L_vs_C['game_result'] == 'W']
wins_count = wins.groupby('fran_id').size().reset_index(name='wins')

plt.figure(figsize=(10, 6))
plt.bar(wins_count['fran_id'], wins_count['wins'], color='pink')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.title('Celtics vs Lakers')
plt.xticks(rotation=45)
plt.show()
```
